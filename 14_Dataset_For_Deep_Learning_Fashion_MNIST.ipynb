## 2.MNIST数据集
MNIST数据集是美国国家标准与技术研究所（National Institute of Standards and Technology）修改后的数据库，是一个著名的手写数字数据集，常用于训练机器学习的图像处理系统。NIST代表国家标准与技术研究所。

M in MNIST代表modified，这是因为有一个原始的NIST数字数据集被修改为MNIST。

MNIST因使用数据集的频率而闻名。 常见的原因有两个：

* 初学者使用它是因为它很容易

* 研究人员使用它来基准化（比较）不同的模型。

数据集由70,000张手写数字图像组成，并分为以下几个部分：

* 60,000张训练图像

* 10,000张测试图像

这些图像最初是由美国人口普查局员工和美国高中生制作的。

MNIST已经被广泛使用，并且图像识别技术已经得到了很大的改进，以至于数据集被认为太简单了。 这就是创建Fashion-MNIST数据集的原因。

## 3. Fashion-MNIST数据集
顾名思义，Fashion-MNIST是时尚商品的数据集。 具体而言，数据集具有以下十种时尚项目类别：

* 0	T-shirt/top
* 1	Trouser
* 2	Pullover
* 3	Dress
* 4	Coat
* 5	Sandal
* 6	Shirt
* 7	Sneaker
* 8	Bag
* 9	Ankle boot

Fashion-MNIST图片来自哪里？ Fashion-MNIST基于Zalando网站上的分类。 Zalando是一家德国跨国时装贸易公司，成立于2008年。

这就是为什么我们在GitHub URL上看到zalandoresearch的原因，其中可以下载Fashion-MNIST数据集。

Zalando Research是创建数据集的公司内部的团队。

当我们回顾介绍数据集的论文时，我们将看到更多有关如何收集图像的信息，但首先，让我们回答另一个潜伏的问题：


Fashion-MNIST数据集之所以具有MNIST的原因，是因为创建者寻求将MNIST替换为Fashion-MNIST。

因此，Fashion数据集被设计为尽可能接近原始MNIST数据集，同时由于与手写图像相比仅具有更复杂的数据而引入了更高的训练难度。

我们将在本文中看到Fashion-MNIST镜像原始数据集的特定方式，但是我们已经看到的一件事是类的数量。

* MNIST –有10个类别（每个数字0-9对应一个）

* FashionMNIST –有10个类别（这是有意的）

阅读[论文](https://arxiv.org/abs/1708.07747)摘要后，我们将了解为什么将数据集命名为Fashion-MNIST。

## 4.Fashion-MNIST是如何建立?

与MNIST数据集不同,Fashion-MNIST不是手工绘制的，而是数据集中的图像是Zalando网站上的实际图像。

但是，它们已被转换为更接近于MNIST规范。 这是每个图像经过的常规转换过程：

* 转换为PNG

* 修整

* 调整大小

* 锐化

* 扩展

* 否定

* 灰阶

要查看此过程的更详细说明，请务必查看[论文](https://arxiv.org/abs/1708.07747)的第二部分。
